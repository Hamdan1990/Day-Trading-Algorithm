```python
#Importing the libraries that we shall use for Data Preprocessing

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as web
import quandl
%matplotlib inline
```

# Importing the Data From Quandl
# To use it, need to have installed Quandl package prior

```python
# Importing American Airlines, United Airlines, and OPEC Crude Oil Prices for the last 10 years
start = '2007-01-01'
end = '2017-01-31'
api_key = "Umpnyms_TdGwsysggyyP"
CrudeOil = quandl.get("OPEC/ORB", authtoken=api_key, start_date=start, end_date=end)
AmericanAirlines = quandl.get("WIKI/AAL", authtoken=api_key, start_date=start, end_date=end)
UnitedAirlines = quandl.get("WIKI/UAL", authtoken=api_key, start_date=start, end_date=end)
```

```python
print("Crude Oil Entries : {}" .format(CrudeOil.shape[0]))
print("American Airlines Entries :{}".format(AmericanAirlines.shape[0]))
print("United AirLines Entries :{} ".format(UnitedAirlines.shape[0]))
print("There seems to uneven entries, needs some Data Cleaning")
```

```python
# We need to remove rows that are not in all three datasets.
# This function removes rows from df1 that are not in df2
def RemoveRows(df1, df2):
    Indexdf1 = df1.index
    Indexdf2 = df2.index
    IndexRemoved = []
    for ind in Indexdf1:
        if ind not in Indexdf2:
            df1.drop(ind, axis  = 0, inplace = True)
            IndexRemoved.append(ind)
    return IndexRemoved
```

```python
# We need to call the above function on all three datasets
# The Reason: Crude Oil is taken from OPEC which is including weekends, whereas United and American Airlines come from 
# the same source so there is uneven row numbers between CrudeOil and United, American
CrudeOilRemovedIndexes = RemoveRows(CrudeOil, AmericanAirlines)
AmericanAirlinesRemovedIndexes= RemoveRows(AmericanAirlines,CrudeOil)
UnitedAirlinesRemovedIndexes = RemoveRows(UnitedAirlines, CrudeOil)
```

```python
print("Crude Oil Entries : {}" .format(CrudeOil.shape[0]))
print("American Airlines Entries :{}".format(AmericanAirlines.shape[0]))
print("United AirLines Entries :{} ".format(UnitedAirlines.shape[0]))
print("All Rows seem to be in order now")
```

```python
# Let's perform some Visualization
UnitedAirlines['Open'].plot(figsize = (16,8), label = 'UAL')
AmericanAirlines['Open'].plot( label = 'AAL')
CrudeOil['Value'].plot(label = 'OIL')
plt.legend()
```

```python
# Creating the Training Set
TrainingSet= pd.concat([AmericanAirlines['Open'],UnitedAirlines['Open'],CrudeOil['Value']], axis = 1)
```

```python
# Rename the Columns of this new DataFrame
TrainingSet.columns = [["AAL",'UAL','OIL']]
```

```python
# Exporting the Training Set to .csv file
TrainingSet.to_csv('Training.csv')
```

```python
#Verifying the Export was correctly done
TS = pd.read_csv('Training.csv', index_col=['Date'], parse_dates=True)
TS.head()
```
